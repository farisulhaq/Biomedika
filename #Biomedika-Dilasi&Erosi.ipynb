```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
path = r'/content/drive/Othercomputers/My Laptop/Materi Kuliah/Semester 6/Biomedika/img/j.png'
img = cv2.imread(path, 0)
img.shape
```

```python
plt.imshow(img)
```

## Kernel

```python
kernel = np.ones((5,5), np.uint8)
```

## Erosi

```python
erosion = cv2.erode(img,kernel,iterations = 1)
erosion.shape
```

```python
plt.imshow(erosion)
```

## Dilasi

```python
dilation = cv2.dilate(img,kernel,iterations = 1)
dilation.shape
```

```python
plt.imshow(dilation)
```
